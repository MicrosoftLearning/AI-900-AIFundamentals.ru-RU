# Диалоговый ИИ

Подумайте о том, как часто вы взаимодействуете с другими людьми с помощью мессенджеров, социальных сетей, электронной почты или других онлайн-технологий. Для многих из нас это основная форма контакта. Если у вас возникает вопрос на работе, вы можете обратиться к коллеге, отправив сообщение в чат, который доступен на мобильном устройстве. Таким образом, вы всегда остаетесь на связи.

![Человек и робот за беседой](./images/conversational_ai.jpg)

Боты – это агенты ИИ, которые используют эти типы каналов для связи, обеспечивая естественное общение с программными службами.

## Создание базы знаний QnA Maker

Для поддержки клиентов часто создается бот, который может интерпретировать и отвечать на часто задаваемые вопросы в окне чата веб-сайта, по эл. почте или через голосовой интерфейс. В основе интерфейса бота лежит база знаний, которая состоит из вопросов и соответствующих ответов. Бот может искать в этой базе знаний подходящие ответы.

Служба QnA Maker представляет собой службу из ряда Cognitive Services, которая позволяет быстро создать базу знаний путем ввода пар «вопрос – ответ» или с использованием существующего документа или веб-страницы. Она может использовать некоторые встроенные возможности обработки естественного языка для интерпретации вопросов и поиска подходящих ответов.

1. Откройте другую вкладку браузера и перейдите на портал QnA Maker по адресу https://qnamaker.ai. Войдите с учетной записью Майкрософт, которая связана с подпиской Azure.
2. На портале QnA Maker выберите **Создать базу знаний**.
3. Если вы ранее не создавали ресурс службы вопросов и ответов, выберите **Создать службу вопросов и ответов**. Портал Azure будет открыт в другой вкладке, чтобы можно было создать службу QnA Maker в рамках вашей подписки. Используйте следующие параметры:
   
   - **Подписка**. *Ваша подписка Azure*
   - **Группа ресурсов**. *Выберите существующую группу ресурсов или создайте новую*
   - **Имя**. *Уникальное имя для ресурса вопросов и ответов*
   - **Ценовая категория**. F0
   - **Ценовая категория службы поиска Azure**. F
   - **Расположение службы поиска Azure**. *Любое доступное расположение*
   - **Имя приложения**. _Не отличается от значения **Имя** (автоматически будет добавлено «.azurewebsites.net»)_
   - **Расположение веб-сайта**. _Не отличается от значения **Расположение службы поиска Azure**_
   - **App insights**. Отключить

> **Примечание**. Если вы уже подготовили бесплатные ресурсы **QnA Maker** или **службу поиска Azure**, ваша квота может не допустить создания еще одного ресурса. В этом случае выберите уровень, отличный от **F0** / **F**.

4. Дождитесь развертывания службы вопросов и ответов и связанных ресурсов для завершения портала Azure.
5. Вернитесь на вкладку портала QnA Maker, затем в разделе **Шаг 2** щелкните **Обновить**, чтобы обновить список доступных ресурсов службы вопросов и ответов.
6. Подключите службу вопросов и ответов к вашей базе знаний, выбрав следующие параметры:
  - **Идентификатор каталога Microsoft Azure**. *Идентификатор каталога Azure для вашей подписки*
  - **Имя подписки Azure**. *Ваша подписка Azure*
  - **Служба вопросов и ответов Azure**. *Ресурс службы вопросов и ответов, созданный на предыдущем шаге*
  - **Язык**. На английском языке

### (!) Проверка
Если отображается сообщение о том, что роль не имеет разрешений на выполнение действия, обновите страницу браузера с порталом QnA Maker.

7. В разделе **Шаг 3** введите имя **База знаний Margie's Travel**.
8. В разделе **Шаг 4** в поле **URL-адрес** введите *https://github.com/MicrosoftDocs/ai-fundamentals/raw/master/data/qna_bot/margies_faq.docx*</a> и щелкните **+ Добавить URL**. Затем в разделе **Беседа** выберите **Профессиональная**.
9. В разделе **Шаг 5** нажмите **Создать базу знаний**.
10. Подождите примерно минуту, пока создается база знаний. Затем просмотрите вопросы и ответы, которые были импортированы из документа часто задаваемых вопросов и предварительно определенные ответы для профессиональной беседы.


## Изменение базы знаний

Ваша база знаний основана на сведениях в документе часто задаваемых вопросов и некоторых предварительно заданных ответов. Вы можете добавить дополнительную пользовательскую пару «вопрос – ответ».

1. Выберите **+ Добавить пару «вопрос – ответ»**.
2. В поле **Вопрос** введите `Hello`. Затем щелкните **+ Добавить альтернативную фразу** и введите `Hi`.
3. В поле **Ответ** введите `Здравствуйте`.

## Обучение и тестирование базы знаний

Теперь у вас есть база знаний и вы можете протестировать ее на портале QnA Maker.

1. В правой верхней части страницы щелкните **Сохранить и обучить** для обучения базы знаний. Может понадобиться изменить размер окна для отображения этой кнопки.
2. По завершении обучения щелкните **&larr; Проверить**, чтобы открыть панель проверки.
3. В нижней части панели проверки введите сообщение *Привет*. Должен вернуться ответ **Здравствуйте**.
4. В нижней части панели проверки введите сообщение *Хочу забронировать авиабилет*. Должен вернуться подходящий ответ из часто задаваемых вопросов.
5. Завершив тестирование базы знаний, нажмите **&rarr; Проверка**, чтобы закрыть панель проверки.


## Создание бота для базы знаний

База знаний представляет собой базовую службу, которую могут использовать клиентские приложения для ответа на вопросы через некий пользовательский интерфейс. Часто этими клиентскими приложениями являются боты. Чтобы база знаний была доступна боту, необходимо опубликовать ее как службу для доступа по HTTP. После этого можно использовать службу Azure Bot для создания и размещения бота, использующего базу знаний для ответа на пользовательские вопросы.

1. В верхней части страницы QnA Maker нажмите **Опубликовать**. Затем на странице **База знаний Margies Travel** нажмите **Опубликовать**.
2. После развертывания службы щелкните **Создать бот**. Это открывает портал Azure на новой вкладке браузера, чтобы можно было создать бот веб-приложения в вашей подписке Azure.
3. На портале Azure создайте бот веб-приложения со следующими настройками (большинство из них будут предварительно заполнены).
  - **Дескриптор бота**. *Уникальное имя вашего бота*
  - **Подписка**. *Ваша подписка Azure*
  - **Группа ресурсов**. *Группа ресурсов, содержащая ресурс QnA Maker*
  - **Расположение**. *То же расположение, что у вашей службы QnA Maker*.
  - **Ценовая категория**. F0
  - **Имя приложения**. *Не отличается от значения **Дескриптор бота**, *.azurewebsites.net* добавляется автоматически
  - **Язык SDK**. *Выберите C# или Node.js*
  - **Ключ авторизации вопросов и ответов**. *Для этого параметра должен быть автоматически задан ключ проверки подлинности для вашей базы знаний вопросов и ответов*
  - **Расположение или план службы приложений.** *Должно задаваться автоматически для подходящего плана и расположения*
  - **Application Insights**. Отключено
  - **Идентификатор и пароль приложения Microsoft**. Автоматическое создание идентификатора и пароля приложения.
4. Дождитесь создания бота (значок уведомления в правом верхнем углу, который выглядит как звонок, будет анимирован во время ожидания). Затем в уведомлении о завершении развертывания щелкните **Перейти к ресурсу** (или альтернативно на главной странице щелкните **Группы ресурсов**, откройте группу ресурсов, где был создан бот веб-приложения, и щелкните его.)
5. На левой панели бота просмотрите **Настройки**, щелкните **Проверить в веб-чате** и подождите, пока бот не отобразить сообщение **Здравствуйте и добро пожаловать!** (Инициализация может занять несколько секунд).
6. Используйте интерфейс проверки чата, чтобы убедиться, что бот отвечает на вопросы из вашей базы знаний в соответствии с ожиданиями. Например, попробуйте отправить *Хочу отменить бронирование гостиницы*.
